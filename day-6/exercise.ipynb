```python
import math
```

```python
# -----------------------------
# Activation Functions
# -----------------------------

def sigmoid(x):
    return 1 / (1 + math.exp(-x))


def relu(x):
    return max(0, x)
```

```python
# -----------------------------
# Single Neuron
# -----------------------------

class Neuron:
    def __init__(self, weights, bias, activation="relu"):
        self.weights = weights
        self.bias = bias
        self.activation = activation

    def forward(self, inputs):
        # Weighted sum
        total = 0
        for w, x in zip(self.weights, inputs):
            total += w * x

        total += self.bias

        # Activation
        if self.activation == "sigmoid":
            return sigmoid(total)
        else:
            return relu(total)
```

```python
# -----------------------------
# Dense Layer
# -----------------------------

class DenseLayer:
    def __init__(self, n_inputs, n_neurons, activation="relu"):
        self.neurons = []

        # Manually initializing weights (no randomness for clarity)
        for i in range(n_neurons):
            weights = [0.1 * (j + 1) for j in range(n_inputs)]
            bias = 0.1
            neuron = Neuron(weights, bias, activation)
            self.neurons.append(neuron)

    def forward(self, inputs):
        outputs = []
        for neuron in self.neurons:
            outputs.append(neuron.forward(inputs))
        return outputs
```

```python
# -----------------------------
# Tiny Neural Network
# -----------------------------

# Layer 1: 3 inputs -> 4 neurons
layer1 = DenseLayer(n_inputs=3, n_neurons=4, activation="relu")

# Layer 2: 4 inputs -> 2 neurons
layer2 = DenseLayer(n_inputs=4, n_neurons=2, activation="sigmoid")
```

```python
# Sample input
inputs = [0.5, -0.3, 0.8]

print("Input:", inputs)

# Forward pass
output_layer1 = layer1.forward(inputs)
print("After Layer 1:", output_layer1)

output_layer2 = layer2.forward(output_layer1)
print("Final Output:", output_layer2)
```
